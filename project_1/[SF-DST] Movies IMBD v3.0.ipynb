# 20. Какая студия сняла больше всего фильмов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Paramount Pictures
3. Columbia Pictures
4. Warner Bros
5. Twentieth Century Fox Film Corporation

```python
import pandas as pd
data = pd.read_csv('data.csv')
companies_set = set(data.production_companies.str.split('|').sum()) #множество студий в data 
companies_df = pd.DataFrame(companies_set, columns = ['Companies']) #преобразуем множество в DataFrame
companies_df['Count'] = list(map(lambda x: data[data.production_companies.str.contains(x,regex=False, na=False)].imdb_id.count(), companies_df['Companies']))

display(companies_df.sort_values(by='Count', ascending = False).head())
```

Так как Universal и Universal Pictures - это одна компания, то эти значения произведенных фильмов необходимо сложить.
Таким образом у студии Universal Pictures: 175 + 185 = 360 фильмов. Это наибольший результат по студиям.

Даже если проверить dataframe по всем дочерним компаниям Warner и Universal

```python
warner = companies_df[companies_df.Companies.str.contains("Warner", na=False)] #Фильтруем по всем компаниям группы Warner
warner_result = warner.append(warner.sum(axis=0), ignore_index=True) #Добавляем в DF сумму фильмов по всем компаниям
display(warner_result)

universal = companies_df[companies_df.Companies.str.contains("Universal", na=False)] #Фильтруем по всем компаниям группы Universal
universal_result = universal.append(universal.sum(axis=0), ignore_index=True) #Добавляем в DF сумму фильмов по всем компаниям
display(universal_result)
```

Группа Universal все равно обладает наибольшим числом произведенных фильмов.
